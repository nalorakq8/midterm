```python
import pandas as pd
import tweepy
import altair as alt
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth',100000)
consumer_key = "**HIDDEN**"
consumer_secret = "**HIDDEN**"
access_token = "**HIDDEN**"
access_token_secret = "**HIDDEN**"

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

# the search will wait if the rate limit is reached and
# contiune after the restriction is removed
api = tweepy.API(auth,wait_on_rate_limit=True,wait_on_rate_limit_notify=True)
```

```python
searched_users = [page for page in tweepy.Cursor(
    api.search_users, q='airlines',count=20).pages(50)]
```

```python
airlines_df = pd.DataFrame()
```

```python
for page in searched_users:
    
    users_list = [
       {
            "user_screen_name":user.screen_name,
            "verified":user.verified,
            "followers_count":user.followers_count,  
            "statuses_count":user.statuses_count,
            "following_count":user.friends_count,
            "location":user.location,
           "profile_created":user.created_at,
           "fav_number":user.favourites_count,
           "description":user.description,
           "user_timezone":user.time_zone,
           "user_name":user.name,
           "user_id":user.id,
           "user_lang":user.lang
        }  for user in page
    ]
    airlines_df = airlines_df.append(pd.DataFrame(users_list),ignore_index=True)
airlines_df.profile_created = pd.to_datetime(airlines_df.profile_created)
airlines_df = airlines_df.drop_duplicates('user_id')
airlines_df=airlines_df.sort_values(by="statuses_count",ascending=False)
airlines_df = airlines_df.reset_index(drop=True)
```

```python
# filtering the verified and english accounts
top_20_airlines_df = airlines_df[(airlines_df.verified==True) & (airlines_df.user_lang == 'en')].reset_index(drop=True).head(39)
top_20_airlines_df.head(21)
```

```python
# removing non-airlines accounts and non-english accounts
top_20_airlines_df = top_20_airlines_df.drop(top_20_airlines_df.index[[7,12,13,20]]).reset_index(drop=True)
```

```python
# now we need to created search query
search_query = ''
for count,screen_name in enumerate(top_20_airlines_df.user_screen_name):
    if count == 19:
        search_query = search_query + 'to:{}'.format(screen_name)
    else:
        search_query = search_query + 'to:{} OR '.format(screen_name)
print(search_query)
```

```python
# searched based on top 20 airlines
searched_tweets = [page for page in tweepy.Cursor(
    api.search, q=search_query,
    lang='en', tweet_mode='extended',count=200).pages(2000)]
```

```python
# storing the results in dataframe
top20_airlines_customers_tweets_df = pd.DataFrame()
for page in searched_tweets:
    
    customers_tweets = [
       {
            "tweet_id":str(tweet.id),
            "text":tweet.full_text,
             "tweet_created_at":tweet.created_at,
            "screen_name":tweet.user.screen_name,
            "verified":tweet.user.verified,
            "followers_count":tweet.user.followers_count,
            "coordinates":tweet.coordinates,
            "statuses_count":tweet.user.statuses_count,
            "following_count":tweet.user.friends_count,
            "location":tweet.user.location,
           "profile_created":tweet.user.created_at,
           "link_color":tweet.user.profile_link_color,
           "user_favourites_count":tweet.user.favourites_count,
           "profileimage":tweet.user.profile_image_url,
           "description":tweet.user.description,
           "sidebar_color":tweet.user.profile_sidebar_fill_color,
           "user_timezone":tweet.user.time_zone,
           "user_name":tweet.user.name,
           "tweet_source":tweet.source,
           "tweet_retweet_count":tweet.retweet_count,
           "tweet_favourites_count":tweet.favorite_count,
           "tweeted_to":tweet.in_reply_to_screen_name

        }  for tweet in page
    ]
    top20_airlines_customers_tweets_df = top20_airlines_customers_tweets_df.append(pd.DataFrame(customers_tweets),ignore_index=True)
top20_airlines_customers_tweets_df.tweet_created_at = pd.to_datetime(top20_airlines_customers_tweets_df.tweet_created_at)
top20_airlines_customers_tweets_df =top20_airlines_customers_tweets_df.sort_values(by="tweet_created_at",ascending=False)
top20_airlines_customers_tweets_df = top20_airlines_customers_tweets_df.reset_index(drop=True)
top20_airlines_customers_tweets_df = top20_airlines_customers_tweets_df.drop_duplicates('tweet_id')
```

```python
# storing top20_airlines_customers_tweets_df dataframe into csv
top20_airlines_customers_tweets_df.to_csv('top20_airlines_customers_tweets.csv',encoding='utf-8',index=False)
```
